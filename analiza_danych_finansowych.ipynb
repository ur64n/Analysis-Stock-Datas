```python
import pandas as pd

# Wczytanie danych z pliku CSV
file_path = r'C:\Users\urban\PycharmProject\Analiza Danych Finansowych\data\stocks.csv'
df = pd.read_csv(file_path)

# Wyświetlenie pierwszych kilku wierszy
print(df.head())
```

```python
# Usunięcie pierwszego wiersza
df = pd.read_csv(file_path, skiprows=1)

# Wyświetlenie pierwszych kilku wierszy
print(df.head())
```

```python
# Sprawdzenie liczby kolumn
print(len(df.columns))
print(df.columns)
```

```python
# Nowe nazwy kolumn po usunięciu pierwszego wiersza
df.columns = ['Date', 'AAPL_Close_Adj', 'GOOGL_Close_Adj', 'MSFT_Close_Adj',
              'AAPL_Close', 'GOOGL_Close', 'MSFT_Close',
              'AAPL_High', 'GOOGL_High', 'MSFT_High',
              'AAPL_Low', 'GOOGL_Low', 'MSFT_Low',
              'AAPL_Open', 'GOOGL_Open', 'MSFT_Open',
              'AAPL_Volume', 'GOOGL_Volume', 'MSFT_Volume']

# Usunięcie ewentualnych spacji w nazwach kolumn
df.columns = df.columns.str.strip()

# Wyświetlenie pierwszych kilku wierszy, aby sprawdzić poprawność
print(df.head())
```

```python
# Podział danych na osobne DataFrame dla każdego tickera
aapl_data = df[['Date', 'AAPL_Close_Adj', 'AAPL_Close', 'AAPL_High', 'AAPL_Low', 'AAPL_Open', 'AAPL_Volume']]
googl_data = df[['Date', 'GOOGL_Close_Adj', 'GOOGL_Close', 'GOOGL_High', 'GOOGL_Low', 'GOOGL_Open', 'GOOGL_Volume']]
msft_data = df[['Date', 'MSFT_Close_Adj', 'MSFT_Close', 'MSFT_High', 'MSFT_Low', 'MSFT_Open', 'MSFT_Volume']]

# Ustawienie kolumny 'Date' jako indeksu
aapl_data.set_index('Date', inplace=True)
googl_data.set_index('Date', inplace=True)
msft_data.set_index('Date', inplace=True)

# Wyświetlenie pierwszych kilku wierszy danych dla AAPL, aby sprawdzić poprawność
print(aapl_data.head())
```

```python
# Usunięcie duplikatów indeksów
aapl_data.index.name = None
googl_data.index.name = None
msft_data.index.name = None

# Wyświetlenie pierwszych kilku wierszy danych dla AAPL, aby sprawdzić poprawność
print(aapl_data.head())
```
